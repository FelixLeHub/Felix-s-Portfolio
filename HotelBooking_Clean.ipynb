```python
# Import libraries
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import dataset to variable df
df = pd.read_csv('hotel_bookings.csv')
df
```

```python
# Remove duplicate columns
duplicate_col = ['Unamed: 0', 'Noname']
df.drop(columns=['Unnamed: 0', 'Noname'], inplace=True)
df
```

```python
# Remove nulls
df.isnull().sum()

# Replace nulls under 'agent'
df['agent'].unique() # Lots of unique values under 'agent'
df[df['agent'] == 1] # 'agent' isn't dependent on other columns, agent number isn't related to other columns
df['agent'].fillna(-1, inplace=True)
df.isna().sum()
```

```python
# Replace nulls under 'children' 
df[df['children'].isnull()] # Only 4 rows, so we drop
df.dropna(subset='children', inplace=True)
df
```

```python
# Replace nulls under 'country'
df['country'].fillna('Unknown', inplace=True)
df['country'].isnull().sum()
```

```python
# Drop unncessary column 'company'
df.drop(columns=['company'], inplace=True)
df.isna().sum()
```

```python
# Standardize column type
# is_canceled --> change to boolean
# children --> change to int64
# is_repeated_guest --> change to boolean
# agent --> chnage to int

df = df.astype({'is_canceled': 'boolean', 'is_repeated_guest':'boolean', 'children':'int', 'agent':'int'})
df.dtypes
```

```python
# Bin lead_time
df['lead_time'].describe()
bin = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0 - 100', '101 - 200', '201 - 300', '301 - 400', '401 - 500', '501 - 600', '601 - 700', '701 - 800']

df['leadtime_binned'] = pd.cut(df['lead_time'], bins=bin, labels=labels)
df[['lead_time', 'leadtime_binned']]
```

```python
# Seperate arrival_time
df ['arrival_month'] = df['arrival_date'].str.split('-', expand=True)[0]
df ['arrival_year'] = df['arrival_date'].str.split('-', expand=True)[1]

column_to_move = df.pop('arrival_month')
df.insert(4, 'arrival_month', column_to_move)

column_to_move = df.pop('arrival_year')
df.insert(5, 'arrival_year', column_to_move)

df.drop(columns=['arrival_date'], inplace=True)

df.head()
```

```python
# String cleaning
df['hotel'] = df['hotel'].replace(r'[\*\n\^]','', regex=True)
df['hotel'].unique()
```

```python
# Remove duplicate rows
df=df.drop_duplicates(keep='first')
df.loc[df.duplicated(keep=False)]
```
